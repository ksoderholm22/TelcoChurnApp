## Data ingest

```python
#function to load data
def load_data(filename) -> pd.DataFrame:
    dir = Path.cwd().parent
    path = dir/'Raw_data'/filename
    print(path)
    if filename.split('.')[1]== 'xlsx':
        data = pd.read_excel(path)
    elif filename.split('.')[1]== 'csv':
        data = pd.read_csv(path)
    else:
        'Check file format'
    return data
```

```python
#Load 5 different excel files 
filename = []
for p in Path( '../Raw_data' ).glob('*.xlsx'):
    print( p.name)
    filename.append(p.name)

for i in range(0,len(filename)):
    exec('{} = load_data(filename[i])'.format(filename[i].split('.')[0]))
    dfname = load_data(filename[i])
    print(filename[i], " loaded in dataframe: ",filename[i].split('.')[0])
    
```

## Join all the datasets to form a master dataset

```python
Telco_customer_churn_population.columns
Telco_customer_churn.columns
Telco_customer_churn_demographics.columns
Telco_customer_churn_services.columns
Telco_customer_churn_status.columns
Telco_customer_churn_location.columns
```

```python
#merge1 = pd.merge(Telco_customer_churn,Telco_customer_churn_demographics, left_on = 'CustomerID',right_on = 'Customer ID', how = 'left')
merge1 = pd.merge(Telco_customer_churn_demographics,Telco_customer_churn_services, on = 'Customer ID', how = 'inner')
merge2 = pd.merge(Telco_customer_churn_status,Telco_customer_churn_location, on = 'Customer ID', how = 'inner')
merge3 = pd.merge(merge1,merge2, on = 'Customer ID', how = 'inner')
all_combined = pd.merge(merge3,Telco_customer_churn_population, on = 'Zip Code', how = 'left')
```
